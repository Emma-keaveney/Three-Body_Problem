#### This notebook is used to visualise the data from the three-body.cpp file

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#.txt file from C++ file saves x,y positions separated by a tab
sat_t = np.loadtxt("satellite_taylor.txt", delimiter="\t")
sat_RK = np.loadtxt("satellite_RK.txt", delimiter="\t")
earth = np.loadtxt("earth.txt", delimiter="\t")
moon = np.loadtxt("moon.txt", delimiter="\t")

plt.figure(figsize = (5,5))

plt.plot(moon[:,0],moon[:,1], label = 'Moon')
plt.plot(earth[:,0],earth[:,1], label = 'Earth')
plt.plot(sat_t[:,0],sat_t[:,1], label = 'Satellite - Taylor')
plt.plot(sat_RK[:,0],sat_RK[:,1], label = 'Satellite - RK')


plt.xlabel('km')
plt.ylabel('km')

plt.legend()
plt.show()
```
